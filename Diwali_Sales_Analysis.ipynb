```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/Diwali Sales Data.csv', encoding='latin-1')
```

```python
df.info()
```

**CHECKING THE NULL VALUES**

```python
pd.isnull(df)
```

**DROP THE COLUMN HAVING NO INSIGHTS**

```python
df.drop(['Status','unnamed1'], axis = 1,inplace = True)
```

```python
pd.isnull(df).sum()
```

**RECHECKING THE NA VALUES**

```python
df.dropna(inplace = True)
```

```python
pd.isnull(df).sum()
```

**CHANGING THE DATA TYPE**

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtype
```

**CHANGING THE NAME OF THE COLUMNS**

```python
df.rename(columns = {'Marital_Status' : 'Married_or_not'})
```

**CHECKING THE NUMERICAL DATA INFORMATION**

```python
df.describe()
```

GETTING INFORMATION OF THE SPECIFIC DATA

```python
df[['Age','Orders','Amount']].describe()
```

**EXPLORATORY DATA ANALYSIS**

```python
df.columns
```

```python
plot_gender = sns.countplot(x = 'Gender', data = df)

for bars in plot_gender.containers:
  plot_gender.bar_label(bars)
```

```python
count_amount = df.groupby(['Gender'], as_index  = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
count_amount
```

```python
sns.barplot(x = 'Gender', y = 'Amount', data = count_amount )
```

from above graph we can clearly see that the apex purchaser are female they have more purchasing power than the man.

STATE WISE order number


```python
df.columns
```

```python
state_wise_order = df.groupby(['State'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(9)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = state_wise_order, x = 'State',y = 'Orders')


```

from above graph we can say that uttar pradesh state has more customers.

```python
ms = sns.countplot(data = df, x = 'Marital_Status')
sns.set(rc={'figure.figsize':(2,5)})
for bars in ms.containers:
  ms.bar_label(bars)
```

```python
gender_marital_status = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plotting_gma = sns.barplot(data=gender_marital_status, x='Marital_Status', y='Amount', hue='Gender')
```

from above graph we can see that the married females have place the most order.

Occupation

```python
sns.set(rc={'figure.figsize':(22,5)})
oc = sns.countplot(data = df, x = 'Occupation')
for bars in oc.containers:
  oc.bar_label(bars)
```

```python
occupation_wise = df.groupby(['Occupation'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False)
sns.set(rc={'figure.figsize':(20,5)})
plot_occu = sns.barplot(data = occupation_wise, x = 'Occupation', y = 'Amount')
```

from above graph we can clerly see that IT sector employee had purchase the most, rank 2 i.e Healthcare and on rank 3. Aviation

Product Category

```python
sns.set(rc = {'figure.figsize':(30,5)})
sns.countplot(data = df, x = 'Product_Category')
```

```python
product_category = df.groupby(['Product_Category'], as_index = False)['Amount'].sum().sort_values(by = 'Amount', ascending = False).head(10)
pc = sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data = product_category, x = 'Product_Category', y = 'Amount')
```

```python
P_id = df.groupby(['Product_ID'], as_index = False)['Orders'].sum().sort_values(by = 'Orders', ascending = False).head(10)
sns.set(rc= {'figure.figsize':(20,5)})
sns.barplot(data = P_id, x = 'Product_ID', y = 'Orders')
```

CONCLUSION

  from all the analysis we see that the married women having age group between 26 - 35 from U.P,Maharashtra and Karnataka working in IT, Healthcare, and Aviation are most like to buy items from food, cloathing and electronic gadjets.
